```python
# https://courses.edx.org/courses/course-v1:MITx+6.00.1x_8+1T2016/courseware/sp13_Week_3/sp13_Problem_Set_3/

# 6.00 Problem Set 3
# 
# Hangman game
#

# -----------------------------------
# Helper code
# You don't need to understand this helper code,
# but you will have to know how to use the functions
# (so be sure to read the docstrings!)

import random
import string

WORDLIST_FILENAME = "words.txt"

def loadWords():
    """
    Returns a list of valid words. Words are strings of lowercase letters.
    
    Depending on the size of the word list, this function may
    take a while to finish.
    """
    print "Loading word list from file..."
    # inFile: file
    inFile = open(WORDLIST_FILENAME, 'r', 0)
    # line: string
    line = inFile.readline()
    # wordlist: list of strings
    wordlist = string.split(line)
    print len(wordlist), "words loaded."
    return wordlist

def chooseWord(wordlist):
    """
    wordlist (list): list of words (strings)

    Returns a word from wordlist at random
    """
    return random.choice(wordlist)

# end of helper code
# -----------------------------------

# Load the list of words into the variable wordlist
# so that it can be accessed from anywhere in the program
wordlist = loadWords()

def isWordGuessed(secretWord, lettersGuessed):
    '''
    secretWord: string, the word the user is guessing
    lettersGuessed: list, what letters have been guessed so far
    returns: boolean, True if all the letters of secretWord are in lettersGuessed;
      False otherwise
    '''
    for char in secretWord:
        if (char in lettersGuessed) == False:
            return False
    return True

def getGuessedWord(secretWord, lettersGuessed):
    '''
    secretWord: string, the word the user is guessing
    lettersGuessed: list, what letters have been guessed so far
    returns: string, comprised of letters and underscores that represents
      what letters in secretWord have been guessed so far.
    '''
    UNDERSCORE = ' _ '
    output = ''
    for char in secretWord:
        if char in lettersGuessed:
            output += char
        else:
            output += UNDERSCORE
    return output



def getAvailableLetters(lettersGuessed):
    '''
    lettersGuessed: list, what letters have been guessed so far
    returns: string, comprised of letters that represents what letters have not
      yet been guessed.
    '''
    ALL_LETTERS = string.ascii_lowercase
    
    output = ''
    
    for char in ALL_LETTERS:
        if (char in lettersGuessed) == False:
            output += ' {0} '.format(char)

    return output
    

def hangman(secretWord):
    '''
    secretWord: string, the secret word to guess.

    Starts up an interactive game of Hangman.

    * At the start of the game, let the user know how many 
      letters the secretWord contains.

    * Ask the user to supply one guess (i.e. letter) per round.

    * The user should receive feedback immediately after each guess 
      about whether their guess appears in the computers word.

    * After each round, you should also display to the user the 
      partially guessed word so far, as well as letters that the 
      user has not yet guessed.

    Follows the other limitations detailed in the problem write-up.
    '''
    
    life_counter = 8
    letters_guessed = []

    def show_welcome_message(secretWord):
        '''
        Print game welcome message
        '''
        print ''
        print 'Welcome to the game, Hangman!'
        print 'I am thinking of a word that is {0} letters long.'.format(len(secretWord))
        print '-------------'
        
    def show_turn_message():
        print 'You have {0} guesses left.'.format(life_counter)
        print 'Available letters:', getAvailableLetters(letters_guessed)
    
    def show_tried_message(letters_guessed):
        print "Oops! You've already guessed that letter:", getGuessedWord(secretWord, letters_guessed)
        print '-------------'
        
    def show_correct_message(lettersGuessed):
        print "Good guess:", getGuessedWord(secretWord, letters_guessed)
        print '-------------'
    
    def show_incorrect_message(lettersGuessed):
        print "Oops! That letter is not in my word:", getGuessedWord(secretWord, letters_guessed)
        print '-------------'
        
    def show_good_game_message():
        print 'Congratulations, you won!'
    
    def show_bad_game_message():
        print 'Sorry, you ran out of guesses. The word was {0}.'.format(secretWord)
    
    def is_ascii(s):
        return all(ord(c) < 128 for c in s)
    
    def is_char_tried(char, letters):
        '''
        Check the char is inside letters_guessed
        '''
        return char in letters
    
    def is_char_in_secretword(char, secretWord):
        '''
        Check the char is inside secretword
        '''
        return char in secretWord
    
    def get_user_input():
        '''
        Get input from user and store the input inside letterGueese
        '''
        user_input = None
        while True:
            user_input = raw_input('Please guess a letter:')
            if is_ascii(user_input) and len(user_input) == 1:
                break
        return user_input
    
    show_welcome_message(secretWord)
    
    print secretWord
    
    while True:
        show_turn_message()
        
        user_input = get_user_input()
        
        if is_char_tried(user_input, letters_guessed):
            show_tried_message(letters_guessed)
        else:
            letters_guessed.append(user_input)
            
            if isWordGuessed(secretWord, letters_guessed):
                show_good_game_message()
                break
            
            if is_char_in_secretword(user_input, secretWord):
                show_correct_message(letters_guessed)
            else:
                life_counter -= 1
                show_incorrect_message(letters_guessed)

        if life_counter == 0:
            break
            
    if isWordGuessed(secretWord, letters_guessed) == False:
        show_bad_game_message()
    
hangman(chooseWord(wordlist))

# When you've completed your hangman function, uncomment these two lines
# and run this file to test! (hint: you might want to pick your own
# secretWord while you're testing)

# secretWord = chooseWord(wordlist).lower()
# hangman(secretWord)
```
