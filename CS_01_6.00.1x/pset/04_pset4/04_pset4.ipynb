```python
# https://courses.edx.org/courses/course-v1:MITx+6.00.1x_8+1T2016/courseware/Week_4/8cb59f02afe3482d9a8cf975abd29f41/

def nfruits(fruit_dictionary, pattern_str):
    '''
    Args:
        fruit_dictionary (dictionary): Non-empty dictionary
        contains type of fruit and its quality
        
        pattern_str (string): A string pattern of the fruits
        eaten by Python on his journey
    
    Returns:
        maximum quantity out of the different types of fruits
        that is available with Python when he has reached the
        campus
    '''
    
    for index in range(len(pattern_str)):
        if index == (len(pattern_str) - 1):
            fruit_dictionary[pattern_str[index]] -= 1
            break
        for key in fruit_dictionary:            
            if key == pattern_str[index]:
                fruit_dictionary[key] -= 1
            else:
                fruit_dictionary[key] += 1
    max = -1
    for key in fruit_dictionary:
        if fruit_dictionary[key] > max:
            max = fruit_dictionary[key]
    return max - 1

nfruits({'A': 5, 'H': 7, 'M': 8, 'N': 9, 'S': 10, 'V': 6}, 'MMM')
nfruits({'U': 7, 'M': 6, 'L': 7}, 'ULLU')
nfruits({'P': 8}, 'P')
```
